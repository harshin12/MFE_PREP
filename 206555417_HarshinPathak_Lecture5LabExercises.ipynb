## 8.3 Pig Latin. LabAssist: Remember strings can be sliced like lists and Arrays

* In this question I converted a "phrase" into pig latin. This was done by first splitting a string into its individual word components. Next, I transformed the words into "piglatin" by adding the first letter to the end of the word and then adding 'ay' to the end of the word. Lastly, I joined all the strings together into a single string.

```python
# input phrase to convert to pig latin

phrase = "harshin is learning about strings"

# create blank list
pig_latin_list = []


# 1) split input phrase into individual component strings
for item in phrase.split(' '):

    # 2) Add the first letter of individual string set (item) to the end of item. Then add 'ay'
    pig_item =  item + item[0] + 'ay'

    # 3) to complete "translation" to pig latin remove the first letter of item
    piglatin_translation = pig_item[1:len(pig_item)]

    # 4) Add all "converted" words to list
    pig_latin_list.append(piglatin_translation)


# 4) Join the list of strings together into a single string
' '.join(pig_latin_list)
```
